```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
intersection_width = np.random.uniform(5,20)
t = np.random.uniform(2,5)
```

```python
class Car:
    def __init__(self,v_zero = (1000/360)*np.random.uniform(20,80), distance_zero = np.random.uniform(10,150),
                 acceleration_positive = np.random.uniform(1,3),deceleration_negative = np.random.uniform(-3,-1)):
        self.v_zero = v_zero 
        self.distance_zero = distance_zero 
        self.acceleration_positive = acceleration_positive
        self.deceleration_negative = deceleration_negative #negative direction
        self.go_mode = True
    def suggest(self):
        go = self.v_zero* t+ 0.5*self.acceleration_positive* t**2
        stop = self.v_zero *t + 0.5* self.deceleration_negative * t**2
        if(go >= self.distance_zero + intersection_width):
            print("go")
            self.go_mode = True
        elif(stop <= self.distance_zero):
            print("stop")
            self.go_mode= False
        else:
            recommended_v_zero = go if abs(go-self.v_zero)< abs(stop-self.v_zero) else stop
            return recomended_v_zero
            print("crush")                   
```

```python
car1 = Car()
car1.suggest()
```

```python
speeds  = range(19,80) #integers from 20 to 80
goings = np.zeros((61,141))
distances = range(9,150)
print(len(speeds))
print(len(distances))
print(goings.shape)
```

```python
for i in range(61):
    for j in range(141):
        car1 = Car(v_zero=(1000/360)*speeds[i],distance_zero = distances[j])
        car1.suggest()
        goings[i,j] = car1.go_mode
#seems like initial speeds dont make much difference
```

```python
plt.imshow(goings, interpolation='none', cmap='gray')
plt.contour(goings, 1, colors='cyan', linewidths=2)
```

we obsert that lowers speeds and highest distances make our car stop.
Black color represents falses in the boolean matrix wher true= go, false = stop
